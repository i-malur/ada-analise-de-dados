IQR -> DISTANCIA INTER QUARTIL
O IQR (Intervalo Interquartil) é uma medida de dispersão estatística que descreve a amplitude dos 50% centrais de um conjunto de dados. Ele é calculado subtraindo o primeiro quartil (Q1) do terceiro quartil (Q3). O IQR é útil para identificar outliers, pois valores que caem fora do intervalo de 1.5 vezes o IQR abaixo de Q1 (outliers para menos) ou acima de Q3 (outiers para mais) são considerados outliers.

```python
q1 = houses['bathroom'].quantile(0.25) # primeiro quartil
q3 = houses['bathroom'].quantile(0.75)
IQR = q3 - q1

print(f'IQR {IQR}')
```

```python
q3
```

```python
houses_outliers = houses[(houses['bathroom'] < (q1 - (1.5 * IQR))) | (houses['bathroom'] > (q3 + (1.5 * IQR)))]
houses_outliers
```

IQR -> 2  
q3 -> 3

3 * (2. 1.5) = 6

Casas com 6 ou mais banheiros são outliers para mais.

```python
# vamos inventer o filtro para pegar os dados que estão dentro da nossa curva
houses_inliers = houses[(houses['bathroom'] >= (q1 - (1.5 * IQR))) & (houses['bathroom'] <= (q3 + (1.5 * IQR)))]
houses_inliers
```
