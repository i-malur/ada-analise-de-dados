Passo a passo deste dataset:

1. Importação de Bibliotecas: As bibliotecas seaborn e matplotlib.pyplot foram importadas para visualização, e pandas para manipulação de dados.
2. Carregamento do Dataset: O conjunto de dados Iris foi carregado diretamente da biblioteca seaborn.
3. Exploração Inicial dos Dados:
  * As primeiras linhas do dataset foram exibidas para ter uma ideia da estrutura dos dados.
  * A contagem de cada espécie no dataset foi verificada, confirmando que o dataset é balanceado (50 amostras por espécie).
  * Estatísticas descritivas das variáveis numéricas foram calculadas para entender a distribuição central, dispersão e amplitude dos dados.
4. Visualização da Distribuição das Variáveis:
  * Histogramas com curva de densidade (KDE) foram gerados para cada variável (sepal_length, sepal_width, petal_length, petal_width) individualmente para visualizar suas distribuições gerais.
  * Histogramas com curva de densidade (KDE) foram gerados para cada variável, mas com a adição da coloração por espécie (hue=iris['species']). Isso permitiu observar como a distribuição de cada variável se comporta dentro de cada espécie.
5. Visualização da Média das Variáveis por Espécie:
  * Gráficos de barras foram gerados para visualizar a média de cada variável (sepal_length, sepal_width, petal_length, petal_width) para cada espécie de íris. A barra de erro (ci=90) indica o intervalo de confiança de 90% para a média.

  Dataset utilizado: 'iris', disponível no seaborn.

```python
import seaborn as sns
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].value_counts()
```

```python
iris.describe()
```

```python
col = 'sepal_length'
sns.histplot(data=iris, x='sepal_length', kde=True).set_title(f'Distribuição da variável {col}')
```

```python
for col in iris.drop(columns='species'):
  sns.histplot(data=iris, x=col, kde=True).set_title(f'\nDistribuição da variável {col}')
  plt.show()
```

```python
for col in iris.drop(columns='species'):
  sns.histplot(data=iris, x=col, kde=True, hue=iris['species']).set_title(f'\nDistribuição da variável {col}')
  plt.show()
```

```python
for col in iris.drop(columns='species'):
  sns.barplot(data=iris, x=iris['species'], hue=iris['species'], y=col, ci = 90)
  plt.show()
```
